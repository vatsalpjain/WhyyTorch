```python
import numpy as np
from Autograd import WhyyTorch as wt
import matplotlib.pyplot as plt
```

```python
"""
What:Creating a Training dataset for the bigram neural network
Why: NN take numbers as input so to predict th next most probable character we need to convert the characters into numbers, and make pairs of characters to train the model on.
"""
words = open('bigram.txt').read().splitlines()
stoi = {'.': 0, **{chr(ord('a') + i): i + 1 for i in range(26)}} 
itos = {i: c for c, i in stoi.items()}
count = 0
print(f"Our Dictionary :{stoi}")
xs,ys = [],[]
for w in words[:]:
    chs = ['.'] + list(w) + ['.']       
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        xs.append(ix1)
        ys.append(ix2)
        count += 1
print(count)
xs = wt(xs)
ys = wt(ys)
```

```python
xs,ys
```

```python
#This is where i convert characters from the dict i made to numbers one hot encoded to go as input to the NN and also he output would be something like this only
xs_one_hot = xs.one_hot(num_classes=27)
ys_one_hot = ys.one_hot(num_classes=27)
print(f"The X: {xs_one_hot} and Shape of X: {xs_one_hot.shape}")
print(f"The Y: {ys_one_hot} and Shape of Y: {ys_one_hot.shape}")
```

```python
"""
What: Initializing the weights of the model(randomly) and this has only 1 parameter called w and shape 27,27
Why: Start with ramdom fhir optimize 
"""
random_seed = 42
np.random.seed(random_seed)
W = wt(np.random.randn(27, 27),requires_grad=True)
```

```python
"""
Do the Forward pass
Step 1: Matrix multiply the one hot encoded input with the weights to get the logits.
Step 2: Apply the exponential function to the logits to get the count.
Step 3: Normalize the counts to get the probabilities.
"""
xs_one_hot,ys_one_hot = xs.one_hot(num_classes=27),ys.one_hot(num_classes=27)
# Logits is raw multiplication of the input with weight , like y=x*w + b but here we dont have b so its just x*w
logits = xs_one_hot @ W 
# exp is used to convert the logits to positive values , so probability can be calculated from it
count = logits.exp() 
# This is the probability of each character , , The last 2 steps together are called softmax function
probs = count / count.sum(axis=1, keepdims=True) 
```

```python
"""
Calculate the Loss
Step 4: Calculate the loss using cross entropy loss function which is -log(probability of the correct class)
"""
#Loss is calculated using cross entropy loss function which is -log(probability of the correct class)
loss = -(ys_one_hot * probs.log()).sum(axis=1).mean()
print(f"The Loss is : {loss}")
```

```python
"""
Optimizer zero grad
Step 5: Set the gradients of the weights to zero
"""
# Optimizer zero grad
W.zero_grad()
```

```python
"""
Loss backwards
Step 6: Backpropagate the loss to calculate the gradients of the weights
"""
loss.backward()
```

```python
"""
Optimizer step
Step 7: Update the weights using the gradients calculated in the previous step and a learning rate
"""
W.data += -0.1 * W.grad
```

```python
# ALL IN ONE
#Forward pass
for i in range(100):
    xs_one_hot,ys_one_hot = xs.one_hot(num_classes=27),ys.one_hot(num_classes=27)
    logits = xs_one_hot @ W 
    count = logits.exp() 
    probs = count / count.sum(axis=1, keepdims=True) 
    #Calculate the loss
    loss = -(ys_one_hot * probs.log()).sum(axis=1).mean()
    print(f"Loss: {loss.data}")
    #Optimizer zero grad
    W.grad = 0.0
    #Loss backwards
    loss.backward()
    #Optimizer step
    W.data += -50 * W.grad
```

```python
# finally, sample from the 'neural net' model
np.random.seed(42)

for i in range(5):
    out = []
    ix = 0
    while True:
    
        xenc   = wt([ix]).one_hot(num_classes=27)
        logits = xenc @ W                              # predict log-counts
        counts = logits.exp()                          # counts, equivalent to N
        p      = counts / counts.sum(axis=1, keepdims=True)  # probabilities for next character
        # ----------

        ix = int(np.random.choice(27, p=p.data[0]))
        out.append(itos[ix])
        if ix == 0:
            break
    print(''.join(out))
```
